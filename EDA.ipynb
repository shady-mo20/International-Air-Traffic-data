# Data: https://www.kaggle.com/datasets/parulpandey/us-international-air-traffic-data

```python
import os
def load_data(file_path):
    _, file_extension = os.path.splitext(file_path)
    if file_extension == '.csv':
        return pd.read_csv(file_path)
    elif file_extension == '.xlsx':
        return pd.read_excel(file_path)
    else:
        raise ValueError("Unsupported file format. Only CSV and XLSX are supported.")

def main():
    data_path = input("Enter the path to the data file (CSV or XLSX): ")
    df = load_data(data_path)

    print("Data loaded successfully. Sample data:")
    print(df.head())

    while True:
        print("\nSelect an option:")
        print("1. Show basic statistics")
        print("2. Plot bar chart of a numeric column over time")
        print("3. Show correlation heatmap")
        print("4. Plot box plot of a numeric column by year")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            print("\nBasic Statistics:")
            print(df.describe())
        elif choice == '2':
            numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
            print("\nNumeric Columns:")
            print(numeric_columns)
            column = input("Enter the name of the numeric column to plot: ")
            plt.figure(figsize=(12, 6))
            sns.barplot(x='Year', y=column, data=df)
            plt.title(f'Average {column} over the years')
            plt.xticks(rotation=45)
            plt.show()
        elif choice == '3':
            plt.figure(figsize=(12, 8))
            sns.heatmap(df.corr(), annot=True)
            plt.title('Correlation Heatmap')
            plt.show()
        elif choice == '4':
            numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
            print("\nNumeric Columns:")
            print(numeric_columns)
            column = input("Enter the name of the numeric column to plot: ")
            plt.figure(figsize=(12, 6))
            sns.boxplot(x='Year', y=column, data=df)
            plt.title(f'Box Plot of {column} by Year')
            plt.xticks(rotation=45)
            plt.show()
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
